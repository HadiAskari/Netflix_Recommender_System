```python
reader = Reader(line_format='user item rating', sep=',', rating_scale=(1, 5))
data = Dataset.load_from_df(training_data1[['customer-id', 'movie-id', 'rating']], reader)

trainset = data.build_full_trainset()

algo = SVD(random_state=40,biased=False)
algo.fit(trainset)
```

```python
a=algo.pu
b=algo.qi
```

```python
print(a.shape)
b=b.T
print(b.shape)
c=np.dot(a,b)
```

```python
np.save("users.npy", c)
```

```python
clusterer = KMeans(n_clusters=5, random_state=10)
cluster_labels = clusterer.fit_predict(c)
```

```python
np.save("clusters.npy",cluster_labels)
```
